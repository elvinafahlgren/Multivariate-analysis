```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
%matplotlib inline
```

```python
fig, ax = plt.subplots()

x = np.linspace(-np.pi, np.pi)
y = np.linspace(-3,3)
X, Y = np.meshgrid(x, y)
Z=Y**2/2-np.cos(X)

ax.contour(X,Y,Z, levels=15)
```
