```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,10))

plt.subplot(221)
x = np.linspace(-2,2)
y = np.linspace(-2,2)
X, Y = np.meshgrid(x, y)
Z1=3*pow(X,2)-5*pow(Y,2)
plt.contour(X,Y,Z1)

plt.subplot(222)
x = np.linspace(-2,2)
y = np.linspace(-2,2)
X, Y = np.meshgrid(x, y)
Z2=3*Y-5*X+1
plt.contour(X,Y,Z2)

plt.subplot(223)
x = np.linspace(-2,2)
y = np.linspace(-2,2)
X, Y = np.meshgrid(x, y)
Z3=3*X-5*Y+1
plt.contour(X,Y,Z3)

plt.subplot(224)
x = np.linspace(-2,2)
y = np.linspace(-2,2)
X, Y = np.meshgrid(x, y)
Z4=3*pow(X,2)+5*pow(Y,2)
plt.contour(X,Y,Z4)

plt.show()
```
