```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff
%matplotlib widget
```

```python
x,y = symbols('x y', real=True)
f = 122 - 16*x + 98*y - 5*x**2 -24*x*y + 5*y**2

#differntiating function f in respect to x
f_diff_x = diff(f, x)
print(f_diff_x)

#differntiating function f in respect to y
f_diff_y = diff(f, y)
print(f_diff_y)
```

```python
def f(x, y):
    return 122 - 16*x + 98*y - 5*x**2 -24*x*y + 5*y**2

def dx(x,y):
    return -10*x - 24*y - 16

def dy(x,y):
    return -24*x + 10*y + 98 

print(f(3,-2))
print(dx(3,-2))
print(dy(3, -2))
```

```python
# L(x,y) = -3 + 2(x-3) + 6(y+2) = 2*x + 6*y -3
```

```python
def L(x, y):
    return 2*x + 6*y -3

x = np.linspace(-2,8)
y = np.linspace(-7,3)

X, Y = np.meshgrid(x, y)
Z1 = L(X, Y)
Z2 = f(X, Y)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z1, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.plot_surface(X, Y, Z2, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```

```python
def E(x, y):
    return np.abs(f(x,y)-L(x,y))

x = np.linspace(2.9,3.1)
y = np.linspace(-2.1,-1.9)

X, Y = np.meshgrid(x, y)
Z = E(X, Y)
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```
