```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10,10))

x = np.linspace(0,4)
y = np.linspace(0,6)
X, Y = np.meshgrid(x, y)
Z=5*np.sin(X)*np.sin(Y/2)-X/3-Y/5
plt.contour(X,Y,Z, 20)

plt.axis("equal")
```

```python
def f(x, y):
    return 5*np.sin(X)*np.sin(Y/2)-X/3-Y/5

x = np.linspace(0,4)
y = np.linspace(0,6)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
```
